```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Sample - Superstore.csv', encoding='latin1')

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
category_sales = df.groupby('Category')['Sales'].sum()

plt.figure(figsize=(8,5))

category_sales.plot(kind='bar')

plt.title('Sales by Category')

plt.ylabel('Sales')

plt.show()
```

```python
profit_category = df.groupby('Category')['Profit'].sum()

plt.figure(figsize=(8,5))

profit_category.plot(kind='bar')

plt.title('Profit by Category')

plt.ylabel('Profit')

plt.show()
```

```python
top_states = df.groupby('State')['Sales'].sum()\
               .sort_values(ascending=False)\
               .head(10)

plt.figure(figsize=(10,5))

top_states.plot(kind='bar')

plt.title('Top 10 States by Sales')

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x='Sales',
    y='Profit'
)

plt.title('Sales vs Profit')

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    df[['Sales','Quantity','Discount','Profit']].corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')

plt.show()
```

# Business Insights

1. Technology category generated the highest sales.

2. Office Supplies had stable sales performance.

3. Some states contribute significantly more revenue than others.

4. High discounts often reduce profit.

5. Sales and Profit show a positive relationship in many orders.

6. Regional performance varies across the business.
